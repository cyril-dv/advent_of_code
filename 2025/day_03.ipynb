```python
import numpy as np
import pandas as pd
np.set_printoptions(suppress=True, precision=4)
pd.options.display.float_format = '{:,.4f}'.format
```

```python
with open('day_03.txt') as f:
    data = f.read()
    data_list = data.split('\n')

data_list[0:1]
```

```python
arr = np.array([[j for j in i] for i in data_list], dtype=np.int32)
arr[0, :]
```

```python
num_len = arr.shape[1]
res = []

for i in range(arr.shape[0]):
    max_val = np.max(arr[i])
    max_pos = np.argmax(arr[i])

    if max_pos == num_len - 2:
        res.append(arr[i][-2:].tolist())
    elif max_pos == num_len - 1:
        max_pos_new = np.argmax(arr[i][:-1])
        n = arr[i][max_pos_new]
        arr[i][max_pos_new] = 0
        max_pos_2 = np.argmax(arr[i][max_pos_new:])
        res.append(np.array([n, arr[i][max_pos_new:][max_pos_2]]).tolist())
    else:
        n = arr[i][max_pos]
        arr[i][max_pos] = 0
        max_pos_new = np.argmax(arr[i][max_pos:])
        res.append(np.array([n, arr[i][max_pos:][max_pos_new]]).tolist())
```

```python
res = np.array(res, dtype='U').tolist()
res = np.array([[i[0] + i[1]] for i in res], dtype=np.int32)

np.sum(res)
```
