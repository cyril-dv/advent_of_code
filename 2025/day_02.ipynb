```python
import numpy as np
import pandas as pd
np.set_printoptions(suppress=True, precision=4)
pd.options.display.float_format = '{:,.4f}'.format
```

```python
with open('day_02.txt') as f:
    data = f.read()
    data_list = data.split(',')

data_list[0:2]
```

```python
arr = np.array([i.split('-') for i in data_list], dtype=np.int64)
arr[0:2]
```

```python
nums = []

for i in range(arr.shape[0]):
    nums.append(np.arange(arr[i, 0], arr[i, 1] + 1).astype('U').tolist())

nums[0][0:3] + nums[0][-3:]
```

```python
ids = []

for n in nums:
    for i in n:
        if len(i) % 2 != 0:
            continue
        else:
            mid = len(i) // 2
            if i[:mid] == i[mid:]:
                ids.append(i)

np.sum(np.array(ids, dtype=np.int64))
```
