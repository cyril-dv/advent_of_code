# Day 1 - Advent of Code 2025

## Part 1

```python
import numpy as np
import pandas as pd
np.set_printoptions(suppress=True, precision=4)
pd.options.display.float_format = '{:,.4f}'.format
```

```python
with open('day_01.txt') as f:
    data = f.read()
    data_list = data.split('\n')

data_list[0:5]
```

```python
nums = np.arange(0, 100)
init_idx = 50

move_dir = [i[0:1] for i in data_list]
move_dir = np.array([int(i.replace('R', '1').replace('L', '-1')) for i in move_dir])

move_amt = np.array([int(i[1:]) for i in data_list])

moves = move_dir * move_amt
moves[0:5]
```

```python
idx = init_idx
zeros = 0

for move in moves:
    idx = (idx + move) % nums.shape[0]
    if nums[idx] == 0:
        zeros += 1

zeros
```

## Part 2

```python
idx = init_idx
crossings = 0

for move in moves:
    start_pos = idx
    next_pos = (start_pos + move) % nums.shape[0]
    dist = abs(move)
    rot = dist // nums.shape[0]

    if start_pos == 0 and next_pos == 0:
        rotation_count = rot

    elif start_pos != 0 and next_pos == 0:
        rotation_count = rot + 1

    elif start_pos == 0 and next_pos != 0:
        rotation_count = rot

    else:
        left_over = (dist % nums.shape[0] if move > 0 else -1 * (dist % nums.shape[0]))
        remaining_dist = start_pos + left_over
        extra = int(remaining_dist < 0 or remaining_dist > 100)
        rotation_count = rot + extra

    crossings += rotation_count
    idx = next_pos

crossings
```
