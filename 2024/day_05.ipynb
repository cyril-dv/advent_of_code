```python
import numpy as np
import pandas as pd
np.set_printoptions(suppress=True, precision=4)
pd.options.display.float_format = '{:,.4f}'.format
```

```python
with open('day_05.txt') as f:
    data = f.read()
```

```python
rules = np.array([i.split('|') for i in data.split('\n\n')[0].split()], dtype=np.int64)
rules[0:3]
```

```python
pages = [row.split(',') for row in data.split('\n\n')[1].split('\n')]
pages = [[int(num) for num in row] for row in pages]
pages[0:3]
```

```python
rules[rules[:, 0] == pages[0][1]][:, 1]
```

```python
middle_pages = []

for page in pages:
    correct_order = 0
    
    for i in range(len(page) - 1):
        remaining_pages = page[i+1:]
        if np.all(np.isin(remaining_pages, rules[rules[:, 0] == page[i]][:, 1])):
            correct_order += 1
    if correct_order == len(page) - 1:
        middle_pages.append(page[len(page)//2])
```

```python
np.sum(middle_pages)
```
